## 3. Business Metrics & Revenue Tracking

Track subscriptions, revenue, and growth.

```python
# Pricing tiers
pricing = {
    "Free": {"price": 0, "requests": 1000, "features": "Basic"},
    "Pro": {"price": 29, "requests": 100000, "features": "Advanced"},
    "Business": {"price": 149, "requests": 1000000, "features": "Full"},
    "Enterprise": {"price": "Custom", "requests": "Unlimited", "features": "Custom"}
}

# Projected customers (conservative estimates)
customer_projections = {
    "Month 1": {"Free": 50, "Pro": 0, "Business": 0, "Enterprise": 0},
    "Month 2": {"Free": 100, "Pro": 2, "Business": 0, "Enterprise": 0},
    "Month 3": {"Free": 200, "Pro": 5, "Business": 0, "Enterprise": 0},
    "Month 6": {"Free": 500, "Pro": 20, "Business": 2, "Enterprise": 0},
    "Month 12": {"Free": 1000, "Pro": 50, "Business": 10, "Enterprise": 1}
}

# Calculate MRR (Monthly Recurring Revenue)
def calculate_mrr(customers):
    mrr = 0
    mrr += customers.get('Pro', 0) * pricing['Pro']['price']
    mrr += customers.get('Business', 0) * pricing['Business']['price']
    # Enterprise assumed at $500/month average
    mrr += customers.get('Enterprise', 0) * 500
    return mrr

# Calculate for each month
revenue_projection = {}
for month, customers in customer_projections.items():
    revenue_projection[month] = {
        "mrr": calculate_mrr(customers),
        "total_customers": sum(customers.values()),
        "paying_customers": sum(v for k, v in customers.items() if k != 'Free')
    }

df_revenue = pd.DataFrame(revenue_projection).T
print("ðŸ’° Revenue Projections:")
print(df_revenue)

# ARR (Annual Recurring Revenue) at Month 12
arr_month_12 = revenue_projection['Month 12']['mrr'] * 12
print(f"\nðŸ“ˆ Projected ARR (Year 1): ${arr_month_12:,}")
```
